## Softmax

```python
def softmax(x):

    c = np.max(x)
    exp_x = np.exp(x - c)
    sum_exp = np.sum(exp_x)
    y = exp_x / sum_exp

    return y
    # return np.exp(x -c) / np.sum(np.exp(x - c))

a = np.array([1010, 1000, 990])
softmax(a)
```

```python
from tensorflow.keras.datasets import mnist
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
```

```python
num_train = Y_train.shape[0]
num_test = Y_test.shape[0]
X_train = X_train.reshape(num_train, -1)
X_test = X_test.reshape(num_test, -1)
X_train.shape, X_test.shape
```

## Initialize network

```python
from neuralnet_mnist_batch import init_network, predict

net = init_network()
```

## Pridict

```python
batch_size = 100
accuracy_cnt = 0

for i in range(0, num_train, batch_size):
    X_batch = X_train[i:i+batch_size, :]
    # print(training_batch.shape)
    Y_batch = predict(net, X_batch)
    # print(pred_batch.shape)
    pred = np.argmax(Y_batch, axis=1)
    accuracy_cnt += np.sum(pred == Y_train[i:i+batch_size])

print("Accuracy:" + str(float(accuracy_cnt) / num_train))
```
